```python
# Converting Data types

df["Pclass"] = df["Pclass"].astype("category")
df["SibSp"] = df["SibSp"].astype("category")
df["Parch"] = df["Parch"].astype("category")
```

```python
df["Survived"] = df["Survived"].astype("category")
```

```python
# Separating categorical column and numerical
df_num = df.select_dtypes(include = "float64")
df_cat = df.select_dtypes(exclude = "float64")
```

```python
df_num["Fare"] = np.log1p(df_num["Fare"])
```

```python
df_num
```

```python
from sklearn.preprocessing import MinMaxScaler
mn = MinMaxScaler()
df_mn = mn.fit_transform(df_num)
```

```python
df_mn
```

```python
df_mn_df = pd.DataFrame(df_mn, columns = df_num.columns, index = df_num.index)
```

```python
df_mn_df
```

```python
sns.displot(df_num["Fare"])
```

```python
# converting categorical to numerical
df_cat.head()
```

```python
df_cat_num = pd.get_dummies(df_cat)
```

```python
df_cat_num.head()
```

```python
# Merging Categorical column and Numerical
df_final = pd.concat([df_cat_num,df_mn_df],axis=1)
```

```python
df_final.head()
```

```python
# Building Model
X = df_final
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test,pred)
sns.heatmap(cm,annot=True)
```
